```python
import pandas as pd
```

```python
data = pd.read_csv("kaggle_bike_competition_train.csv", header=0, error_bad_lines=False)
data.head()
```

```python
temp = pd.DatetimeIndex(data['datetime'])
data['date'] = temp.date
data['time'] = temp.time
```

```python
data['hour'] = pd.to_datetime(data.time, format="%H:%M:%S")
data['hour'] = pd.Index(data['hour']).hour
```

```python
data['dayofweek'] = pd.DatetimeIndex(data.date).dayofweek
```

```python
data['days'] = (data.date - data.date[0]).astype('timedelta64[D]')
```

```python
CASUAL = data['casual'].groupby(data['dayofweek']).mean()
CASUAL.reset_index()
```

```python
REG = data['registered'].groupby(data['dayofweek']).mean()
```

```python
REG.reset_index()
```

```python
data['sat'] = 0
data['sun'] = 0
data.sat[data['dayofweek'] == 5] = 1
data.sun[data['dayofweek'] == 6] = 1
data.head()
```

```python
dataRel = data.drop(['datetime', 'count', 'date', 'time', 'dayofweek'], axis=1)
```

```python
dataRel
```

```python
from sklearn.feature_extraction import DictVectorizer
```

```python
featureConCols = ['temp','atemp','humidity','windspeed','days','hour']
```

```python
dataFeatureCon = dataRel[featureConCols]
dataFeatureCon = dataFeatureCon.fillna( 'NA' ) #in case I missed any
X_dict_con  = dataFeatureCon.T.to_dict().values()
```

```python
vec = DictVectorizer(sparse = False)
X_con = vec.fit_transform(X_dict_con)
X_con
```
